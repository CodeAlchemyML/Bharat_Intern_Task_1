**Importing required libraries:**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

**Gathering required data:**

```python
dataset = pd.read_csv( '/content/housing.csv' )
dataset.dropna( inplace = True )
print( dataset.head() )
```

**Data splitting and preprocessing:**

```python
x = dataset[ 'total_rooms' ].values
y = dataset[ 'median_house_value' ].values
x = x.reshape( len( x ), 1 )
y = y.reshape( len( y ), 1 )
```

```python
x_train = x[ : 14447 ] #First 14448 values will be the training data
x_test = x[ 14447 : ]  #Rest values will be the testing data
y_train = y[ : 14447 ] #First 14448 values will be the training data
y_test = y[ 14447 : ]  #Rest values will be the testing data
```

**Model building, training and testing:**

```python
regr_model = LinearRegression()
regr_model.fit( x_train, y_train )
plt.scatter( x_test, y_test, label = 'Actual datapoints', color = 'green' )
plt.plot( x_test, regr_model.predict( x_test ), label = 'Predicted datapoints', color = 'orange' )
plt.xlabel( 'Total_no, of rooms' )
plt.ylabel( 'Median house value' )
plt.legend( loc = 'upper right' )
plt.title( 'Plot of median house value w.r.t. total no, of rooms' )
plt.show()
```

**Using the model for predictions:**

```python
new_data = [ [2], [1], [45], [67] ]
print( regr_model.predict( new_data ) )
```
